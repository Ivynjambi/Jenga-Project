```python
prices = pd.read_csv("foods.csv")
```

Since for the item labels to be visible the plot size needs to be enlarged, therefore I've changed the default size to be width = 17 and height = 5.

```python
plt.rcParams["figure.figsize"] = [17, 5]
```

Below is a scatter plot of the prices of different foods during the first recorded month, that is January 2017 and the last recorded month, that is April 2020.

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.scatter(prices["1-Apr-20"], prices["1-Jan-17"])
plt.xlabel("Food Prices on 1st April, 2020")
plt.ylabel("Food Prices on 1st January 2017")
plt.title("Scatter plot of the food prices in 1st January 2017 and 1st April 2020")
plt.show()
```

From the above scatter plot of the last month recorded and the first month recorded;
We can see that most food items are priced below 500 Naira.
We can also see a few outliers in the data, passed the items priced above 1500 Naira.
Overall, there is a strong, positive, linear association between the prices in the first month and the prices in the last month.

We can also print out the food items whose prices are above 1500 Naira between Jan 2017 and Apr 2020

```python
prices[(prices["1-Jan-17"]>1500)& (prices["1-Apr-20"]>1500)]
```

We find that the food items that are constantly above 1500 Naira from 2017 to 2020 are F17 and F32, that is 'Catfish, dried' and 'Mudfish, dried' respectively.

Since the code above shows the relationship between the food items and their prices through out 2017-2020, which is too compact and hard to read, we will get and average of the prices through out 2017 to 2020, then plot the food items against the average prices.

```python
av_row = prices.mean(axis=1)
print(av_row)
```

Below is a bar plot of the average food prices from 2017-2020, against the food item labels

```python
#av_row.plot()
#fig
plt.bar(prices['Item Labels'], av_row)
plt.xlabel("Item labels")
plt.ylabel("Average of food prices from 2017-2020")
plt.title("Bar plot of the average food prices of different foods in Nigeria.")
plt.show()
```

From above, we can see that item F10-'Agric eggs, medium size (price of one)' has the lowest average price, which makes sense since it is a record of the price of each Agric egg. Right after we have item F40-'Sweet potato' which has the next lowest average price recorded between 2017-2020.
The highest average priced item is item F32-'Mudfish, dried' followed right after by item F17-'Catfish, dried'.


Below is a histogram of the average of the food item prices through 2017-2020

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.ylabel("Frequency")
plt.xlabel("Average prices")
plt.title("Average prices of food items distribution in 2017-2020")
hist = av_row.hist()
```

From the figure above we can tell how the data (Average prices of the food items from 2017-2020) is distributed.
The figure represents a right-skewed distribution, which means the mean has the largest value of the measures of central tendancy, which is located to the right of the distribution, and the mode is the smallest value, which is located to the left of the distribution.
