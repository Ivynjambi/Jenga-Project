```python
prices = pd.read_csv("foods.csv")
```

```python
prices.plot()
plt.xlabel("Item Labels")
plt.ylabel("Food Prices")
plt.show()
```

Since the data is too compact and hard to read, we will get and average of the prices through out 2017 to 2020, then plot the food against the average

```python
av_row = prices.mean(axis=1)
print(av_row)
```

```python
#av_row.plot()
plt.plot(prices['Item Labels'], av_row)
plt.xlabel("Item labels")
plt.ylabel("Average of food prices from 2017-2020")
plt.show()
```
