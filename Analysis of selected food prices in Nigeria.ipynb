Next, visualizing the data.

We need to shorten the item labels.
The CSV file has been modified, as F(n-1) with 'F' standing for food, followed by the number (n-1). The food's position in the original list is (n-1). For reference of the original item labels view below.

As mentioned above, the item labels are shortened to F(n), the new csv file is saved as 'foods'.

```python
prices = pd.read_csv("foods.csv")
```

A few plots to visualize the data.

```python
#prices.plot()
#plt.xlabel("Item Labels")
#plt.ylabel("Food Prices")
#plt.show()
```

The code above is a line plot that shows the relationship between the food items and their prices through out Jan 2017- April 2020, which is too compact and hard to read. We will get an average of the prices through out 2017 to 2020, then plot the food items against the average prices.

Below we get the average price for each of the 41 food items

```python
av_row = prices.mean(axis=1)
print(av_row)
```

Below is a bar plot of the food item labels against the average food prices from 2017-2020.

```python
plt.rcParams["figure.figsize"] = [17, 5]
plt.bar(prices['Item Labels'], av_row)
plt.xlabel("Item labels")
plt.ylabel("Average of food prices from 2017-2020")
plt.title("Bar plot of the average food prices of different foods in Nigeria.")
plt.show()
```

From above, we can see that item F10 - 'Agric eggs, medium size (price of one)' has the lowest average price, which makes sense since it is a record of the price of each Agric egg. Right after we have item F40 - 'Sweet potato' which has the next lowest average price recorded between 2017-2020.
The highest average priced item is item F32 - 'Mudfish, dried' followed right after by item F17 - 'Catfish, dried'.


We can look into more different plots to understand the data better.

Below is a scatter plot of the prices of different food items during April 2017 and April 2020.

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.scatter(prices["1-Apr-20"], prices["1-Apr-17"])
plt.xlabel("Food Prices on 1st April, 2020")
plt.ylabel("Food Prices on 1st April, 2017")
plt.title("Scatter plot of the food prices in 1st April 2017 and 1st April 2020")
plt.show()
```

From the above scatter plot of the prices recorded on April 2017 and April 2020;
We can see that for different food items there is not much change between their prices on April 2017 and April 2020, that is 3 years later.
We can also see a few outliers in the data, among the food items priced above 1500 Naira.
Overall, there is a strong, positive, linear association between the prices in April 2017 and in April 2020.

We can also print out the food items whose prices are above 1500 Naira from April 2017 to April 2020

```python
prices[(prices["1-Apr-17"]>1500)& (prices["1-Apr-20"]>1500)]
```

We find that the food items that are constantly above 1500 Naira from April 2017 to 2020 are F17, F25 and F32, that is 'Catfish, dried','Frozen chicken' and 'Mudfish, dried' respectively.

Below is a histogram of the average of the food item prices through 2017-2020, so as to know more about the distribution of the data set.

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.ylabel("Frequency")
plt.xlabel("Average prices")
plt.title("Histogram of the Average prices of food items from 2017-2020")
hist = av_row.hist()
```

From the figure above we can tell how the data (Average prices of the food items from 2017-2020) is distributed.
The figure represents a right-skewed distribution, which means the mean has the largest value of the measures of central tendancy, which is located to the right of the distribution, and the mode is the smallest value of the measures of central tendancy, which is located to the left of the distribution.

We can also look into plotting data yearly.

Since the data under 'prices' is per month, it's combined on excel to have the prices recorded yearly, and the new csv file is 'foods-yr'.

```python
df = pd.read_csv("foods-yr.csv")
```

We take a look at the data by viewing the first 5 columns, with the column 'years' as the DataFrame's index

```python
df = df.set_index("Years")
df.head(5)
```

We look at some line graphs.

Below are line graphs of different food items that are named in the key, showing their prices from 2017 to 2020, allowing us to view if there are any visible patterns.

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.plot(df["F1"], label= "Bread, sliced 500g")
plt.plot(df["F3"], label= "Gari yellow, sold loose")
plt.plot(df["F16"], label= "Broken rice (Ofada)")

plt.xlabel("Years")
plt.ylabel("Food Prices")
plt.title("Prices of different food items from 2017-2020")
plt.legend()
plt.show()
```

Above we see the visualization of the prices of sliced bread, gari(yellow, sold loose) and broken rice(Ofada) from 2017-2020.
Initially for the price of bread the price rises till 2018, then a drastic drop all through the year to 2019, while for the other two items, the prices drops from 2017 to 2019. Then the price begins to rapidly increase for all three items all through to April, 2020, where the price of sliced bread, gari yellow and broken rice is at it's highest, that is, 316.55 Naira, 211.83 Naira and 390.54 Naira respectively.

```python
plt.rcParams["figure.figsize"] = [10, 5]
plt.plot(df["F13"], label= "Beef, bone in")
plt.plot(df["F30"], label= "Maize grain, white sold loose")
plt.plot(df["F43"], label= "Vegetable oil, 1 bottle, specify bottle")

plt.xlabel("Years")
plt.ylabel("Food Prices")
plt.title("Prices of different food items from 2017-2020")
plt.legend()
plt.show()
```

Above we see the visualization of the prices of beef(bone in), maize grain(white, sold loose) and vegetable oil(1 bottle), from 2017 - 2020. For the beef(bone in) the price rapidly drops from 2017 to 2018, then it slightly drops from 2018 to 2019 and then the price begins to rise from 2019 to 2020. For both maize grain(white, sold loose) and vegetable oil(1 bottle) the prices slightly increase from 2017 to 2018, then they rapidly drop from 2018 to 2019, and finally, the price of maize grain(white, sold loose) steadily increases from 2019 to 2020, while the price of vegetable oil(1 bottle) rapidly increases from 2019 to 2020. These final prices are 1047.12 Naira, 551.92 Naira, 165.83 Naira for the beef(bone in), vegetable oil(1 bottle) and maize grain(white, sold loose), respectively.

### Conclusion

What could have caused this rapid increase in food prices from 2019?
There are a number of factors to consider and one of them is the ban on dollar access. The Nigerian president ordered the central bank to stop providing foreign exchange for food and fertilizer imports, in an effort to boost local farming production. In turn the prices of food have increased and are likely to continue increasing with the pandemic faced through 2020-2021.
